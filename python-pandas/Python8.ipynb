```python
def profession_code(s: str):
    code: int = -1
    
    if(s=="Рабочий" or s == "Менеджер"):
        code = 0 if s=="Рабочий" else 1
    else:
        code = 2
    
    return code
        
def age_category(age):
    answ=["молодой", "средний", "зрелый"]
    ind = 0
    
    if age > 35:
        ind = 2
    elif age < 23:
        code = 0
    else:
        ind = 1
    
    return answ[ind]    
```
